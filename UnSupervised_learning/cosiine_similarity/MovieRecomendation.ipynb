#movie recomendation system on the basis of the popularity

#using the cosine similarty algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# used to compare sequences—mostly strings, lines of text, or files. and also used to find how mush close two sequences are.
import difflib

# to convert a collection of raw documents to a matrix of TF-IDF features.
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df =pd.read_csv("../../dataset/movies.csv")
df.head()
```

```python
# no.of rows and colum in dataset
df.shape
```

```python
# selecting the relevent feature from the dataset choosing only 5 features

features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
# replacing the nul values with null  string
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
# combining all 5 features together

combined_features = df['genres']+ ' ' + df['keywords'] + ' ' + df['tagline'] + ' ' + df['cast'] + ' ' + df['director']
combined_features.head()
```

```python
# coneverting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
feature_vectors.shape
print(feature_vectors)
```

```python
# getting the simmilarity scores using the cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# # getting the user input and find the similar type of movies
movie_name = "thor"
```

```python
# creating  a list of all the movies name given in the dataset
list_of_all_movies = df['title'].tolist()
print(list_of_all_movies)
```

```python
# finding the close match for the movie name given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_movies)

close_match = find_close_match[0]
print(find_close_match[0])
```

```python
# find the index of the movie from its title
index_of_movie =df[ df.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting the list of the similar moviess
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# finding the highest similar movies values
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
#  getting the top 10 similar movies
print("Movies suggested for you:\n")
i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['title'].values[0] 
    if(i<30):
        print(i, '.', title_from_index)
        i+=1
```

#Movie Recomendation system


```python
movie_name = "iron man"
# creating  a list of all the movies name given in the dataset
list_of_all_movies = df['title'].tolist()
# finding the close match for the movie name given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_movies)
close_match = find_close_match[0]
# find the index of the movie from its title
index_of_movie =df[ df.title == close_match]['index'].values[0]

# getting the list of the similar moviess
similarity_score = list(enumerate(similarity[index_of_movie]))

# finding the highest similar movies values
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)

#  getting the top 10 similar movies
print("Movies suggested for you:\n")
i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['title'].values[0] 
    if(i<21):
        print(i, '.', title_from_index)
        i+=1
```
